```python

import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt

# Question 1(1.2): Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
tesla_data.head()

#1.3 
import pandas as pd
import yfinance as yf

tesla = yf.Ticker("TSLA")

tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)

tesla_financials = tesla.financials.T
tesla_financials.reset_index(inplace=True)

tesla_revenue = tesla_financials[['index', 'Total Revenue']].rename(columns={'index': 'Date', 'Total Revenue': 'Revenue'})

print(tesla_revenue.head())


#1.4
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)
gme_data.head()


#1.5
import pandas as pd
import yfinance as yf

gme = yf.Ticker("GME")

gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)

gme_financials = gme.financials.T
gme_financials.reset_index(inplace=True)

gme_revenue = gme_financials[['index', 'Total Revenue']].rename(columns={'index': 'Date', 'Total Revenue': 'Revenue'})

gme_revenue.dropna(subset=['Revenue'], inplace=True)
gme_revenue['Revenue'] = gme_revenue['Revenue'].astype(int)

print(gme_revenue.tail())


#1.6
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 1, figsize=(10, 8))

axes[0].plot(tesla_data["Date"], tesla_data["Close"])
axes[0].set_title("Tesla Stock Price")
axes[0].set_xlabel("Date")
axes[0].set_ylabel("Price ($)")

axes[1].plot(tesla_revenue["Date"], tesla_revenue["Revenue"].astype(float))
axes[1].set_title("Tesla Revenue")
axes[1].set_xlabel("Date")
axes[1].set_ylabel("Revenue (Millions $)")

plt.tight_layout()
plt.show()


#1.7
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

axes[0].plot(gme_data["Date"], gme_data["Close"])
axes[0].set_title("GameStop Stock Price")
axes[0].set_xlabel("Date")
axes[0].set_ylabel("Price ($)")

axes[1].plot(gme_revenue["Date"], gme_revenue["Revenue"].astype(float))
axes[1].set_title("GameStop Revenue")
axes[1].set_xlabel("Date")
axes[1].set_ylabel("Revenue (Millions $)")

plt.tight_layout()
plt.show()
```
